## 학습 내용
- Grid
- Block
- thread

Grid는 Block으로 이뤄짐, Block은 thread로 이뤄짐

다음 코드에서의 Layout
- 하나의 블록을 이루는 thread의 개수 : 3개
- 하나의 그리드를 이루는 block의 개수 : 2개

다음 두 함수는 각각의 idx에 위치 값을 알 수 있다
- threadIdx
- blockIdx

다음 두 함수는 가각의 dim값을 알 수 있다
- blockDim
- gridDim

=> 접근하고자 하는 thread에 idx로 접근 가능

```python
%%writefile helloCUDA.cu
#include <stdio.h>
#include <stdlib.h>
#include <cuda_runtime.h>

__global__ void checkIndex(void) {
    printf("threadIdx : (%d %d %d) blockIdx : (%d %d %d) blockDim : (%d %d %d) gridDim : (%d %d %d)\n",
    threadIdx.x, threadIdx.y, threadIdx.z,
    blockIdx.x, blockIdx.y, blockIdx.z,
    blockDim.x, blockDim.y, blockDim.z,
    gridDim.x, gridDim.y, gridDim.z);
}

int main() {
    int n = 6;

    dim3 block(3);
    dim3 grid((n + block.x - 1) / block.x);

    printf("grid.x %d grid.y %d grid.z %d\n", grid.x, grid.y, grid.z);
    printf("block.x %d block.y %d block.z %d\n", block.x, block.y, block.z);

    checkIndex<<<grid, block>>>();
    cudaDeviceReset();
    return 0;
}
```

```python
!nvcc helloCUDA.cu -o helloCUDA
```
